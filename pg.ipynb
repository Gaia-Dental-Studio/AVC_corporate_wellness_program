### School Outreach

```python
import pandas as pd
import numpy as np
```

```python
school_outreach_prices = pd.read_csv('school_outreach_data/treatment_prices_new.csv')
school_treatment = pd.read_csv('school_treatment.csv')
treatment_details = pd.read_csv('treatment_with_details.csv')
```

```python
school_merged = pd.merge(
    school_treatment.assign(lower_treatment=school_treatment['Standardized Treatment'].str.lower()),
    treatment_details.assign(lower_treatment=treatment_details['Treatment'].str.lower()),
    how='left',
    left_on='lower_treatment',
    right_on='lower_treatment'
).drop(columns=['lower_treatment'])

school_merged
```

```python
# drop treatment y, and rename treatment x toTtreatment 
school_merged = school_merged.drop(columns=['Treatment_y'])
school_merged = school_merged.rename(columns={'Treatment_x': 'Treatment'})
print(school_merged)
```

```python
school_prices = pd.read_csv('school_outreach_data/treatment_prices_new.csv')
school_prices
```

```python
school_prices_gaia_IDR = school_prices.copy()

print(school_prices)
```

```python
# Save the original case of 'Treatment' and 'Category' columns
school_prices_gaia_IDR['Original_Treatment'] = school_prices_gaia_IDR['Treatment']
school_prices_gaia_IDR['Original_Category'] = school_prices_gaia_IDR['Category']

school_merged['Original_Treatment'] = school_merged['Treatment']
school_merged['Original_Category'] = school_merged['Category']

# Ensure consistent formatting for 'Treatment' and 'Category' in both DataFrames (lowercase and stripped)
school_prices_gaia_IDR['Treatment'] = school_prices_gaia_IDR['Treatment'].str.strip().str.lower()
school_prices_gaia_IDR['Category'] = school_prices_gaia_IDR['Category'].str.strip().str.lower()

school_merged['Treatment'] = school_merged['Treatment'].str.strip().str.lower()
school_merged['Category'] = school_merged['Category'].str.strip().str.lower()

# Create combined keys in both DataFrames after formatting
school_prices_gaia_IDR['Treatment_Category'] = school_prices_gaia_IDR['Treatment'] + '_' + school_prices_gaia_IDR['Category']
school_merged['Treatment_Category'] = school_merged['Treatment'] + '_' + school_merged['Category']

# Perform the mapping using the combined key
school_prices_gaia_IDR['Original Price (Rp.)'] = school_prices_gaia_IDR['Treatment_Category'].map(school_merged.set_index('Treatment_Category')['total_price_IDR'])
school_prices_gaia_IDR['Cost Material (Rp.)'] = school_prices_gaia_IDR['Treatment_Category'].map(school_merged.set_index('Treatment_Category')['total_cost_IDR'])
school_prices_gaia_IDR['total_duration'] = school_prices_gaia_IDR['Treatment_Category'].map(school_merged.set_index('Treatment_Category')['total_duration'])

# Restore the original case for 'Treatment' and 'Category' from the saved columns
school_prices_gaia_IDR['Treatment'] = school_prices_gaia_IDR['Original_Treatment']
school_prices_gaia_IDR['Category'] = school_prices_gaia_IDR['Original_Category']

school_merged['Treatment'] = school_merged['Original_Treatment']
school_merged['Category'] = school_merged['Original_Category']

# Drop the temporary columns if no longer needed
school_prices_gaia_IDR.drop(columns=['Treatment_Category', 'Original_Treatment', 'Original_Category'], inplace=True)
school_merged.drop(columns=['Treatment_Category', 'Original_Treatment', 'Original_Category'], inplace=True)
```

### Dentist Fee per Hour

```python
# in school_prices, make list out of the 'Treatment' column
school_prices = pd.read_csv('school_outreach_data/treatment_prices_new.csv')
treatment_list = school_prices['Treatment'].tolist()
treatment_list
```

```python
# from treatment_list create a dataframe called 'Dentist Fee per Hour' and please add this value rule: if it is consultation, periapical x-ray, filling, scalling then it is 35000 else it will be 160000
Dentist_Fee_per_Hour = pd.DataFrame(treatment_list, columns=['Treatment'])
Dentist_Fee_per_Hour['Dentist Fee per Hour'] = np.where(Dentist_Fee_per_Hour['Treatment'].isin(['Consultation', 'Periapical X-Ray', 'Filling', 'Scaling']), 35000, 160000)
Dentist_Fee_per_Hour
```

```python
# now load the school_outreach_data/treatment_prices_GAIA_IDR.csv
school_prices_gaia_IDR = pd.read_csv('school_outreach_data/treatment_prices_GAIA_IDR.csv')

Dentist_Fee_per_Hour_unique = Dentist_Fee_per_Hour.groupby('Treatment').first().reset_index()

# now add column 'Dentist Fee per Hour' to school_prices_gaia_IDR
school_prices_gaia_IDR['Dentist Fee per Hour'] = school_prices_gaia_IDR['Treatment'].map(Dentist_Fee_per_Hour_unique.set_index('Treatment')['Dentist Fee per Hour'])
```

```python
# do the same as above for 'school_outreach_data/treatment_prices_new.csv' 
school_prices = pd.read_csv('school_outreach_data/treatment_prices_new.csv')
school_prices['Dentist Fee per Hour'] = school_prices['Treatment'].map(Dentist_Fee_per_Hour_unique.set_index('Treatment')['Dentist Fee per Hour'])
```

```python
school_prices.drop(columns=['Unnamed: 0'], inplace=True)
school_prices_gaia_IDR.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# remove the index column
school_prices.to_csv('school_outreach_data/treatment_prices_new.csv', index=False)
school_prices_gaia_IDR.to_csv('school_outreach_data/treatment_prices_GAIA_IDR.csv', index=False)
```

```python
denominator = 100
random_volumes = np.random.randint(1, denominator, size=11)
random_volumes
```

```python
random_volumes.sum()
```

```python
final_volume = max(denominator - np.sum(random_volumes), 1)
final_volume
```

```python
# Append the final volume
volumes = np.append(random_volumes, final_volume)
volumes
```

```python

# Scale the volumes to ensure they sum up to the exact total
volumes = volumes / np.sum(volumes) * denominator
volumes
```
